On the book website, www.statlearning.com, there is a gene expression data set (Ch12Ex13.csv) that consists of 40 tissue samples with measurements on 1,000 genes. The first 20 samples are from healthy patients, while the second 20 are from a diseased group.

(a) Load in the data using `pd.read_csv()`. You will need to select `header=None`.

(b) Apply hierarchical clustering to the samples using correlation-based distance, and plot the dendrogram. Do the genes separate the samples into the two groups? Do your results depend on the type of linkage used?

(c) Your collaborator wants to know which genes differ the most across the two groups. Suggest a way to answer this question, and apply it here.

```python
!pip install ISLP
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from ISLP import load_data
from scipy.cluster.hierarchy import \
     (dendrogram,
      cut_tree)
from ISLP.cluster import compute_linkage
from sklearn.cluster import \
     (KMeans,
      AgglomerativeClustering)
```

```python
from scipy.stats import ttest_ind
```

```python
gene = pd.read_csv('/content/Ch12Ex13.csv', header=None).transpose()
gene.head()
```

```python
scaler = StandardScaler(with_std=True,
                        with_mean=True)
gene_scaled = scaler.fit_transform(gene)
labels = np.concatenate([np.zeros(20), np.ones(20)])
```

```python
HClust = AgglomerativeClustering

def plot_gene(linkage, ax, cut=-np.inf):
    cargs = {'above_threshold_color':'black',
             'color_threshold':cut}
    hc = HClust(n_clusters=None,
                metric='correlation',
                distance_threshold=0,
                linkage=linkage.lower()).fit(gene_scaled)
    linkage_ = compute_linkage(hc)
    dendrogram(linkage_,
               ax=ax,
               labels=labels,
               leaf_font_size=10,
               **cargs)
    ax.set_title('%s Linkage' % linkage)
    return hc
```

```python
fig, axes = plt.subplots(1, 3, figsize=(30,15))
ax = axes[0]; hc_comp = plot_gene('Complete', ax)
ax = axes[1]; hc_avg = plot_gene('Average', ax)
ax = axes[2]; hc_sing = plot_gene('Single', ax)
```

```python
linkage_comp = compute_linkage(hc_comp)
comp_cut = cut_tree(linkage_comp, n_clusters=2).reshape(-1)
pd.crosstab(pd.Series(labels, name='labels'),
            pd.Series(comp_cut.reshape(-1), name='Complete'))
```

```python
fig, ax = plt.subplots(figsize=(8,6))
plot_gene('Complete', ax, cut=1.12)
ax.axhline(1.12, c='r', linewidth=3);
```

- **Distinct Clusters** : Samples labeled as 0 (`Healthy`) and 1 `(Diseased`) cluster together distinctly, the genes effectively separate the groups.
- Results consistent over the different types of linkages.

```python
healthy = gene[:20]
diseased = gene[20:]
t_stats, p_values = ttest_ind(healthy, diseased, axis=1)
significant_genes = np.argsort(p_values)[:10]  # Top 10 genes with the most significant differences

top_genes_df = pd.DataFrame({
    'Gene': significant_genes,
    'p-value': p_values[significant_genes]
})
print("Top 10 genes with the most significant differences:")
print(top_genes_df)
```

- **t-test**: Compares the means of two independent samples (healthy vs. diseased) for each gene.
- **p-values**: Indicate the probability that the observed differences are due to chance. Smaller p-values suggest more significant differences.
- **Significant Genes**: By sorting the genes based on p-values, we can identify those that differ the most between the two groups.
